```python
batting_first_winner=df4[df4['win_by_runs']!=0]
batting_first_winner
```

```python
batting_first_winner.head(10)
```

```python
plt.figure(figsize=(15,5))
plt.hist(batting_first_winner['win_by_runs'])
plt.title('Distribution of runs')
plt.xlabel('Runs')
plt.show()
```

```python
batting_first_winner['winner']
```

```python
batting_first_winner['winner'].value_counts()
```

```python
plt.figure(figsize=(15,5))
plt.bar(list(batting_first_winner['winner'].value_counts().keys()),list(batting_first_winner['winner'].value_counts()))
plt.xticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(25,25))
plt.pie(list(batting_first_winner['winner'].value_counts()),labels=list(batting_first_winner['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()             
```
