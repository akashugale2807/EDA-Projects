```python
df1['HrSecondary_Girls']=df1['HrSecondary_Girls'].clip(lower=df1['HrSecondary_Girls'].quantile(0.05), upper=df1['HrSecondary_Girls'].quantile(0.95))
df1['HrSecondary_Boys']=df1['HrSecondary_Boys'].clip(lower=df1['HrSecondary_Boys'].quantile(0.05), upper=df1['HrSecondary_Boys'].quantile(0.95))
df1['Primary_Girls']=df1['Primary_Girls'].clip(lower=df1['Primary_Girls'].quantile(0.05), upper=df1['Primary_Girls'].quantile(0.95))
df1['Primary_Boys']=df1['Primary_Boys'].clip(lower=df1['Primary_Boys'].quantile(0.05), upper=df1['Primary_Boys'].quantile(0.95))
df1['Upper Primary_Girls']=df1['Upper Primary_Girls'].clip(lower=df1['Upper Primary_Girls'].quantile(0.05), upper=df1['Upper Primary_Girls'].quantile(0.95))
df1['Upper Primary_Boys']=df1['Upper Primary_Boys'].clip(lower=df1['Upper Primary_Boys'].quantile(0.05), upper=df1['Upper Primary_Boys'].quantile(0.95))
df1['Secondary _Girls']=df1['Secondary _Girls'].clip(lower=df1['Secondary _Girls'].quantile(0.05), upper=df1['Secondary _Girls'].quantile(0.95))
df1['Secondary _Boys']=df1['Secondary _Boys'].clip(lower=df1['Secondary _Boys'].quantile(0.05), upper=df1['Secondary _Boys'].quantile(0.95))
```
